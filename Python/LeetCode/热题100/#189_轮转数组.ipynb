# 解答

```python
class Solution:
    def rotate(self, nums: list[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        for i in range(k):
            for j in range(n - 1):
                nums.append(nums[j])
            for j in range(n - 1):
                nums.pop(0)
```

# 官方

方法一：使用额外的数组

```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        nums_new = nums.copy()
        for i in range(n):
            nums[(i + k) % n] = nums_new[i]
```
